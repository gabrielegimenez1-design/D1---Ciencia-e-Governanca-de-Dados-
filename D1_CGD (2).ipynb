```python
# Importando as bibliotecas que vou utilizar ao longo da análise.
# O pandas e o numpy são usados para manipular e tratar os dados.
# Já o matplotlib e o seaborn vão me ajudar a criar os gráficos e visualizações.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Aqui configuro o estilo dos gráficos para deixá-los com uma aparência mais limpa e visualmente agradável.
plt.style.use('seaborn-v0_8')
```

```python
# Nesta etapa será feita a leitura da base de dados que contém informações
# sobre PIB per capita, taxa de homicídios e taxa de mortalidade.
# Esses dados foram organizados em um arquivo CSV.
base_geral = pd.read_csv("base_geral.csv")

# Exibo as primeiras linhas para ter uma visão inicial das colunas e dos dados disponíveis.
base_geral.head()
```

```python
# Agora que os dados estão limpos, começa as análises estatísticas básicas.
# Isso me ajuda a entender a distribuição, média, mediana e dispersão de cada variável.
# Essas medidas são essenciais para ter uma noção inicial dos padrões do conjunto de dados.
base_geral.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionar apenas colunas numéricas da base real
colunas_numericas = ['PIB_per_capita_R$', 'Taxa_Homicidios_100k', 'Taxa_Mortalidade_1000']

# Calcular correlação apenas entre essas colunas
correlacoes = base_geral[colunas_numericas].corr()

# Criar o mapa de calor
plt.figure(figsize=(8,6))
sns.heatmap(correlacoes, annot=True, cmap="Blues", fmt=".2f")
plt.title("Mapa de Correlação entre Indicadores (PIB, Homicídios e Mortalidade)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gráfico 1 — Relação entre PIB per capita e Taxa de Homicídios
plt.figure(figsize=(8,6))
sns.scatterplot(data=base_geral, x='PIB_per_capita', y='Taxa_Homicidios')
plt.title('Relação entre PIB per capita e Taxa de Homicídios')
plt.xlabel('PIB per capita (R$)')
plt.ylabel('Taxa de homicídios (por 100 mil habitantes)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from mpl_toolkits.mplot3d import Axes3D  # para o gráfico 3D

# Configuração visual e criação da pasta de saída

plt.style.use('seaborn-v0_8')
sns.set_palette("Set2")

# Carregamento da base de dados

base_geral = pd.read_csv("base_geral.csv")

# Padronizar nomes das colunas (caso necessário)
base_geral.rename(columns={
    'pib_per_capita_2022_R$': 'PIB_per_capita_R$',
    'homicides_per_100k_est_2023': 'Taxa_Homicidios_100k',
    'mortality_per_1000_est': 'Taxa_Mortalidade_1000'
}, inplace=True)


# Verificação de dados ausentes

print("Valores ausentes por coluna:")
print(base_geral.isnull().sum())

# Remover registros incompletos
base_geral = base_geral.dropna(subset=['PIB_per_capita_R$', 'Taxa_Homicidios_100k', 'Taxa_Mortalidade_1000'])


# VISUALIZAÇÕES COMPARATIVAS


# Pairplot (todas as combinações possíveis entre as variáveis)

sns.pairplot(base_geral[['PIB_per_capita_R$', 'Taxa_Homicidios_100k', 'Taxa_Mortalidade_1000']], diag_kind='kde')
plt.suptitle('Comparações entre PIB, Homicídios e Mortalidade', y=1.02)
plt.savefig("imagens/pairplot_comparacoes.png", dpi=300, bbox_inches='tight')
plt.show()

# Regressões lineares comparativas

plt.figure(figsize=(8,6))
sns.regplot(data=base_geral, x='PIB_per_capita_R$', y='Taxa_Homicidios_100k', scatter_kws={'alpha':0.7})
plt.title('Tendência: PIB per capita x Taxa de Homicídios')
plt.xlabel('PIB per capita (R$)')
plt.ylabel('Taxa de homicídios (por 100 mil hab.)')
plt.savefig("imagens/reg_PIB_Homicidios.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(8,6))
sns.regplot(data=base_geral, x='PIB_per_capita_R$', y='Taxa_Mortalidade_1000', scatter_kws={'alpha':0.7}, color='green')
plt.title('Tendência: PIB per capita x Taxa de Mortalidade')
plt.xlabel('PIB per capita (R$)')
plt.ylabel('Taxa de Mortalidade (por 1.000 hab.)')
plt.savefig("imagens/reg_PIB_Mortalidade.png", dpi=300, bbox_inches='tight')
plt.show()


# Gráfico 3D - PIB, Homicídios e Mortalidade

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    base_geral['PIB_per_capita_R$'],
    base_geral['Taxa_Homicidios_100k'],
    base_geral['Taxa_Mortalidade_1000'],
    c=base_geral['Taxa_Homicidios_100k'],
    cmap='coolwarm',
    s=60
)

ax.set_title('Relação Tridimensional: PIB x Homicídios x Mortalidade')
ax.set_xlabel('PIB per capita (R$)')
ax.set_ylabel('Taxa de Homicídios (100k)')
ax.set_zlabel('Taxa de Mortalidade (1000)')
plt.savefig("imagens/3D_PIB_Homicidios_Mortalidade.png", dpi=300, bbox_inches='tight')
plt.show()


# Heatmap de Correlação

colunas_numericas = ['PIB_per_capita_R$', 'Taxa_Homicidios_100k', 'Taxa_Mortalidade_1000']
correlacoes = base_geral[colunas_numericas].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlação entre PIB, Homicídios e Mortalidade')
plt.savefig("imagens/heatmap_correlacao.png", dpi=300, bbox_inches='tight')
plt.show()
```
